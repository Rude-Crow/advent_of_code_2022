```python
with open("input_day_6.txt") as input_string:
    input_list = []
    input_string = list(input_string)[0]
    for i in input_string:
        input_list.append(i)

marker = 4
buffer_list = []
for i in range(len(input_list) - (marker - 1)):
    buffer = input_list[i:i+marker]
    buffer = list(dict.fromkeys(buffer))
    if len(buffer) == marker:
        buffer_list.append([buffer, i+marker])


print(buffer_list[0])
```
